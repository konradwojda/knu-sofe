```python
from Slicer_normal import Slicer
import inspect
DATA_TRACKER = "_data"

def demo(x: int) -> int:
    z = x
    while x <= z <= 64:
        z *= 2
    return z


if __name__ == "__main__":
    with Slicer() as slicer:
        demo(10)
    
    _, start_demo = inspect.getsourcelines(demo)
    g = slicer.dependencies().backward_slice(('z', (demo, start_demo + 1))).graph()
```
